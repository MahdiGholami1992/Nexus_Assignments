### Task 3 ‚Äî Load ride log


```python
# üëâ # TODO: read "rides.csv" into df
# (columns: date,temp,rides,weekday)
import pandas as pd
dates = pd.date_range(start="2025-07-01", periods=365)
temps = np.random.normal(loc=20, scale=5, size=365).round(1)
rides = np.random.randint(100, 500, size=365)
weekdays = dates.day_name()
df = pd.DataFrame({
    "date": dates,
    "temp": temps,
    "rides": rides,
    "weekday": weekdays
})
print(df.head())
df.to_csv('/content/sample_data/rides.csv')


df = pd.read_csv("/content/sample_data/rides.csv")
print(df.head())
```

### Task 4 ‚Äî Weekday averages

```python
# üëâ # TODO: compute and print mean rides per weekday
mean_rides = df.groupby("weekday")["rides"].mean()
print(mean_rides)
```

### Task 5 ‚Äî Scatter plot

```python
# üëâ # TODO: scatter-plot temperature (x) vs rides (y) from df

import matplotlib.pyplot as plt

plt.scatter(df["temp"], df["rides"], alpha=0.6)
```

### Task 6 ‚Äî Show the figure



```python
# üëâ # TODO: call plt.show() so the plot appears

import matplotlib.pyplot as plt

plt.scatter(df["temp"], df["rides"], alpha=0.6)
plt.xlabel("Temperature (¬∞C)")
plt.ylabel("Number of Rides")
plt.title("Temperature vs Rides")
plt.grid(True)
plt.show()
```
